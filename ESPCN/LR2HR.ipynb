```python
from utils import *
import tensorflow as tf
import os, cv2
import numpy as np
from PIL import Image
```

```python
gpus = tf.config.experimental.list_physical_devices(device_type='GPU')
print("find gpus:",gpus)
for gpu in gpus:
    tf.config.experimental.set_memory_growth(gpu, True)
```

```python
batch = 1
inh = 90
inw = 65
inc = 3
espcn = ESPCN(4)
espcn.build(input_shape=(batch,inh,inw,inc))
espcn.load_weights("./model/ESPCN")
```

```python
temp_dir = "./temp_sr/"
if bool(1-os.path.exists(temp_dir)):
    os.mkdir(temp_dir)
    print("temp_dir not exists, create one.")

lr_list = os.listdir("../data/cut_data")
lr_list.sort(key=lambda x:int(x[5:-4]))

for i in range(len(lr_list)):
    path = os.path.join("../data/cut_data",lr_list[i])
    img = Image.open(path)
    inputs = np.array(img)
    inputs = np.float32(inputs[np.newaxis,...]/255)
    pred = espcn.predict(inputs)[0]
    pred = Image.fromarray(np.uint8(255*pred),"RGB")
    pred.save(os.path.join(temp_dir,lr_list[i]))
    print("progress %s" % lr_list[i])
```

```python
# INTER_LINEAR
temp_dir = "./temp_INTER_LINEAR/"
if bool(1-os.path.exists(temp_dir)):
    os.mkdir(temp_dir)
    print("temp_dir not exists, create one.")

lr_list = os.listdir("../data/cut_data")
lr_list.sort(key=lambda x:int(x[5:-4]))

for i in range(len(lr_list)):
    path = os.path.join("../data/cut_data",lr_list[i])
    img = Image.open(path)
    img_arr = np.array(img)/255
    h,w,d = np.shape(img_arr)
    pred = cv2.resize(img_arr,(w*4,h*4),interpolation=cv2.INTER_LINEAR)
    pred = Image.fromarray(np.uint8(255*pred),"RGB")
    pred.save(os.path.join(temp_dir,lr_list[i]))
    print("progress %s" % lr_list[i])
```
