```python
from utils import *
import tensorflow as tf
tf.config.experimental_run_functions_eagerly(True)
import os, cv2
import numpy as np
from PIL import Image
```

```python
# # Forbidden GPUs
# os.environ["CUDA_DEVICE_ORDER"] = "PCI_BUS_ID"
# os.environ["CUDA_VISIBLE_DEVICES"] = "-1"

gpus = tf.config.experimental.list_physical_devices(device_type='GPU')
print("find gpus:",gpus)
for gpu in gpus:
    tf.config.experimental.set_memory_growth(gpu, True)
```

```python
batch = 10
inh = 180
inw = 390
inc = 3
upscale = 2
```

```python
# def customized_loss(y_true, y_pred):
#     return tf.reduce_sum(tf.reduce_mean(tf.square(y_pred - y_true)))

fsr = FSR(upscale)
fsr.build(input_shape=(batch,inh,inw,inc))
print(fsr.summary())

fsr.load_weights("./model/fsr2x")
```

```python
temp_dir = "./temp_sr/"
if bool(1-os.path.exists(temp_dir)):
    os.mkdir(temp_dir)
    print("temp_dir not exists, create one.")

lr_list = os.listdir("../data/540cut")
lr_list.sort(key=lambda x:int(x[5:-4]))

for i in range(len(lr_list)):
    path = os.path.join("../data/540cut",lr_list[i])
    img = Image.open(path)
    inputs = np.array(img)
    inputs = np.float32(inputs[np.newaxis,...])
    pred = fsr.predict(inputs)[0]
    pred = Image.fromarray(np.uint8(pred),"RGB")
    pred.save(os.path.join(temp_dir,lr_list[i]))
    print("progress %s" % lr_list[i])
```

```python
# INTER_LINEAR
temp_dir = "./temp_INTER_LINEAR/"
if bool(1-os.path.exists(temp_dir)):
    os.mkdir(temp_dir)
    print("temp_dir not exists, create one.")

lr_list = os.listdir("../data/540cut")
lr_list.sort(key=lambda x:int(x[5:-4]))

for i in range(len(lr_list)):
    path = os.path.join("../data/540cut",lr_list[i])
    img = Image.open(path)
    img_arr = np.array(img)/255
    h,w,d = np.shape(img_arr)
    pred = cv2.resize(img_arr,(w*upscale,h*upscale),interpolation=cv2.INTER_LINEAR)
    pred = Image.fromarray(np.uint8(255*pred),"RGB")
    pred.save(os.path.join(temp_dir,lr_list[i]))
    print("progress %s" % lr_list[i])
```
